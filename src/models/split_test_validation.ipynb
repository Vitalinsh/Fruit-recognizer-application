```python
import shutil
import os

import numpy as np
import random
```

```python
data_dir = 'dataset/test'
test_dir = 'dataset/test_new'
val_dir = 'dataset/validation'

test_data_portion = 0.5
val_data_portion = 0.5
```

```python
classes = os.listdir(data_dir)
n_classes = len(classes)
print(n_classes)
print(classes[:5])
```

```python
def create_directory(dir_name, classes_list):
    
    if os.path.exists(dir_name):
        shutil.rmtree(dir_name)
    os.makedirs(dir_name)
    
    for class_name in classes_list:   
        os.makedirs(os.path.join(dir_name, class_name))
```

```python
random.seed(5)
```

```python
for class_name in classes:
    image_list = os.listdir(os.path.join(data_dir, class_name))
    random.shuffle(image_list)
    
    n_valid_img = len(image_list) // 2
    n_test_img = len(image_list) - n_valid_img
    
    valid_list = image_list[: n_valid_img]
    test_list = image_list[n_valid_img:]
    
    for img in valid_list:
        shutil.copy2(os.path.join(data_dir, class_name, img), 
                    os.path.join(val_dir, class_name))
        
    for img in test_list:
        shutil.copy2(os.path.join(data_dir, class_name, img), 
                    os.path.join(test_dir, class_name))
        
    print("\n", class_name, ":")
    print("total number of imgs =", len(image_list))
    print("number of imgs for valid =", n_valid_img)
    print("number of imgs for test =", n_test_img)
```
